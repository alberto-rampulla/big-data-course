## Classificazione binaria

Da questo momento il codice si frammenta e si eseguiranno i tre codici separatamente. In questa esercitazione porremo focus sulla variabile target `death` che rappresenta un outcome binario:
- 0 vivo;
- 1 morto

Per evitare data leakage informativo rimuoviamo le feature prognostiche che possono dare indicazioni di come un'osservazione debba essere classificata.
 - `aps`, `sps` score clinici;
 - `dzgroup`, `dzclass` gruppi diagnostici;

```python
diag_to_drop = ['aps', 'sps', 'dzgroup', 'dzclass']
dd = dd.drop(columns = diag_to_drop)
dd.shape
```

Come nell'esempio precedente, rimuoviamo le variabili prognostiche esistenti e, dopo aver effettuato lo splitting, eseguiamo tutto il pre-processing necessario : rimozioni sparse -> imputazione -> encoding -> standardizzazione -> analisi correlazione (2 step) -> feature selection.

1. Procedere allo split train-test secondo il rapporto 95%-5% in forma stratificata secondo i valori della variabile target.

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
```

```python
X = dd.drop(columns = ['death'])
y = dd['death']
```

```python
X_tr, X_te, y_tr, y_te = train_test_split(X, 
                                          y, 
                                          test_size = 0.05, 
                                          stratify = y, 
                                          random_state = 42)
```

```python
imp_num = SimpleImputer(strategy='median')
imp_cat = SimpleImputer(strategy='constant', fill_value='missing')
```

```python
num_cols = X_tr.select_dtypes(include = 'number').columns
num_cols
```

```python
cat_cols = X_tr.select_dtypes(exclude = 'number').columns
cat_cols
```

```python
X_tr[cat_cols].apply(lambda x: x.unique())
```

```python
ord_cols = ['income', 'ca']
```

```python
# Imputazione
X_tr[num_cols] = imp_num.fit_transform(X_tr[num_cols])
X_tr[cat_cols] = imp_cat.fit_transform(X_tr[cat_cols])


X_te[num_cols] = imp_num.transform(X_te[num_cols])
X_te[cat_cols] = imp_cat.transform(X_te[cat_cols])
```

```python
X_tr.isna().mean().sort_values(ascending = False)
```

```python
X_te.isna().mean().sort_values(ascending = False)
```

Sono stati imputati correttamente i dati mancanti. Si procede con l'ordinal encoding.

```python
ord_mods = [['missing', 'under $11k', '$11-$25k', '$25-$50k', '>$50k'], ['missing', 'no', 'yes', 'metastatic']]

enc1 = OrdinalEncoder()
enc2 = OrdinalEncoder(categories = ord_mods)
```

```python
new_cat_cols = cat_cols.drop(ord_cols)
X_tr[new_cat_cols] = enc1.fit_transform(X_tr[new_cat_cols])
X_tr[ord_cols] = enc2.fit_transform(X_tr[ord_cols])

X_te[new_cat_cols] = enc1.transform(X_te[new_cat_cols])
X_te[ord_cols] = enc2.transform(X_te[ord_cols])
```

```python
sc = StandardScaler()

X_tr = sc.fit_transform(X_tr)
X_te = sc.transform(X_te)
```

In questo caso, a differenza del precedente, in cui veniva richiesto il clustering, adesso è necessario valutare anche la correlazione tra la risposta e le features, rimuovendo quelle che assumono valori troppo bassi e suggeriscono di non essere rulevanti per prevedere i valori di y.

```python
soglia = 0.1

col_to_remove = (
    X_tr.corrwith(y_tr)
    .abs()
    .sort_values(ascending = False)
    .loc[lambda x: x < soglia]
    .index
)

len(col_to_remove) # 25 features poco correlate con il target
```

```python
X_tr = X_tr.drop(columns = col_to_remove)
X_te = X_te.drop(columns = col_to_remove)
```

```python
soglia = 0.8
coppie = (
    X_tr.corr()
    .abs()
    .map(lambda x: x > soglia)
    .unstack()
    .reset_index()
)

coppie.columns = ['var1', 'var2', 'high_corr']

(coppie 
 .loc[lambda x: (x['high_corr'] == True) & (x['var1'] < x['var2']),
      ['var1', 'var2']
      ])
```

```python
X_tr['survmean'] = X_tr[['surv2m', 'surv6m']].mean(axis = 1)
X_te['survmean'] = X_te[['surv2m', 'surv6m']].mean(axis = 1)

X_tr['prgmean'] = X_tr[['prg2m', 'prg6m']].mean(axis = 1)
X_te['prgmean'] = X_te[['prg2m', 'prg6m']].mean(axis = 1)
```

```python
X_tr = X_tr.drop(columns = ['surv2m', 'surv6m', 'prg2m', 'prg6m'])
X_te = X_te.drop(columns = ['surv2m', 'surv6m', 'prg2m', 'prg6m'])
```

```python
X_tr.shape
```

```python
X_te.shape
```

È stata calcolata una variabile sintetica combinazione lineare delle variabili multicollineari, data dalla **media per riga delle feature interessate**; successivamente sono state rimosse quelle originali.

```python
from sklearn.feature_selection import SelectKBest, f_classif
fs = SelectKBest(score_func=f_classif, k = "all").fit(X_tr, y_tr)
fs.pvalues_
```

```python
skb = SelectKBest(score_func = f_classif, k = 4)
X_tr_skb = skb.fit_transform(X_tr, y_tr)
X_te_skb = skb.transform(X_te)
```

```python
X_tr_skb.columns.to_list()
```

## Classificatore binario

### RandomForestClassifier / LogisticRegression / SVC / AdaBoost

Utilizzare RandomForestClassifier per la classificazione con i seguenti iperparametri:
1. criterion: "gini", "log_loss";
2. min_samples_split: 2, 5, 10;
3. max_features: "sqrt", 5

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
rf = RandomForestClassifier(random_state = 42)
param_grid = {
    'criterion': ['gini', 'log_loss'],
    'min_samples_split': [2, 5, 10, 20],
    'max_features': ['sqrt', 5]
}
grid_rf = GridSearchCV(estimator = rf, param_grid = param_grid, cv = 5, n_jobs = -1)
grid_rf.fit(X_tr_skb, y_tr)
print("Miglior modello:", grid_rf.best_params_)

best_rf = grid_rf.best_estimator_
y_pred = best_rf.predict(X_te_skb)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_te, y_pred))
```

```python
# Curva ROC e relativa AUC
from sklearn.metrics import RocCurveDisplay
auc_disp = RocCurveDisplay.from_estimator(best_rf, X_te_skb, y_te)
```

Il modello proposto è contraddistinto da una buona AUC pari a 0.79; il modello identificato dalla combinazione di iperparametri scelta è un buon modello.

### Logistic Regression


```python
# Esempio alternativo con regressione logistica
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver = 'liblinear', random_state = 42)
# Il parametro C controlla la forza della regolarizzazione, più è piccolo più è forte
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2']
}
grid_lr = GridSearchCV(estimator = lr, param_grid = param_grid, cv = 5, n_jobs = -1)
grid_lr.fit(X_tr_skb, y_tr)
print("Miglior modello:", grid_lr.best_params_)

best_lr = grid_lr.best_estimator_
y_pred_lr = best_lr.predict(X_te_skb)
```

```python
print(classification_report(y_te, y_pred_lr))
```

```python
auc_disp_lr = RocCurveDisplay.from_estimator(best_lr, X_te_skb, y_te)
```

In termini di AUC, il modello di regressione logistica selezionato dalla griglia di iperparametri ha evidenziato una migliore performance di quello precedentemente mostrato.

### Support Vector Classifier

```python
# Esempio alternativo con SVM: il support vector classifier
from sklearn.svm import SVC
svc = SVC(random_state= 42)
param_grid = {
    'kernel': ['linear', 'rbf'], #, 'poly' per kernel polinomiale
    'C': [0.1, 1, 10],
    'gamma': ['scale', 'auto'],
    # 'degree': [3, 4, 5] #solo per kernel polinomiale
}

grid_svc = GridSearchCV(estimator = svc, param_grid = param_grid, cv = 5, n_jobs = -1)
grid_svc.fit(X_tr_skb, y_tr)
print("Miglior modello:", grid_svc.best_params_)
best_svc = grid_svc.best_estimator_
y_pred_svc = best_svc.predict(X_te_skb)
```

```python
print(classification_report(y_te, y_pred_svc))
```

```python
auc_disp_svc = RocCurveDisplay.from_estimator(best_svc, X_te_skb, y_te)
```

Il modello con SVC è migliore rispetto al RandomForest, ma peggiore della regressione logistica in termini di AUC.

### AdaBoost

```python
# Esempio alternativo con AdaBoost
from sklearn.ensemble import AdaBoostClassifier
adb = AdaBoostClassifier(random_state = 42)
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.1, 1]
}
grid_adb = GridSearchCV(estimator = adb, param_grid = param_grid, cv = 5, n_jobs = -1)
grid_adb.fit(X_tr_skb, y_tr)
print("Miglior modello:", grid_adb.best_params_)

best_adb = grid_adb.best_estimator_
y_pred_adb = best_adb.predict(X_te_skb)
```

```python
print(classification_report(y_te, y_pred_adb))
```

```python
auc_disp_adb = RocCurveDisplay.from_estimator(best_adb, X_te_skb, y_te)
```

Performance simile al modello logistico, in generale i modelli più complessi non sempre performano meglio di quelli più semplici
